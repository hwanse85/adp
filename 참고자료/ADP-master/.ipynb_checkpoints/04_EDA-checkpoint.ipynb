## 전처리

- 임시 로딩 : row가 몇 개인지, 데이터 타입이 뭔지 모르기 때문

```python
df = pd.read_csv('ade_sales.csv', nrows = 1000)
```

- 중복행 처리

```python
df.head()
df.tail()
df.describe() # count 수가 다른 것이 있는지(=결측값)
df.duplicated().sum() # 중복행 여부 확인
df[df.duplicated()] # 중복행 확인
df.drop_duplicates(inplace = True) # 중복행 제거
df = df.reset_index(drop = True) 
# 행 제거 후에는 reset_index 해 주는 게 좋음
```

- 날짜 처리

```python
df.dtypes
df['Date'] = pd.to_datetime(df['Date'])
```

- NaN 처리

```python
df.isna()
df.isna().any() # 열별 nan 보유여부 확인
df.isna().any(1) # 행별 nan 보유여부 확인
df.isna().sum() # 컬럼별 nan 개수 확인
df.isna().any(1).nonzero() # nan이 있는 행 index 찾기 : 여기서는 20행
```

```python
df.dropna(inplace = True) # nan 포함 행 삭제
df.fillna(0, inplace = True) # nan을 0으로 대체
df.fillna(method = 'ffill') # 직전값으로 채우기
df.fillna(method = 'bfill') #직후값으로 채우기
df.loc[19,'Leaflets'] = df.loc[17:22'Leaflets'].mean() 
# 부분 평균값으로 채우기

df.loc[19,'Leaflets'] = df['Leaflets'].mean() # 평균값으로 
df.loc[19,'Leaflets'] = df['Leaflets'].median # 중앙값으로
```

- 이상치 처리

```python
ti = pd.read_csv('titanic_simple.csv')
```

```python
sns.boxplot(y = 'Fare', data = ti, orient = 'h') ; plt.show()
```

```python
ti['Fare'].describe()
```

```python
sns.distplot(ti[ti['Fare'].notna()]['Fare'], rug = True) ; plt.show()
```

```python
# IQR 활용
q1 = ti['Fare'].quantile(.25)
q3 = ti['Fare'].quantile(.75)
iqr = q3-q1
min_iqr = q1 - 1.5 * iqr
max_iqr = q3 + 1.5 * iqr
min_from_all = ti['Fare'].min()
max_from_all = ti['Fare'].max()
if (min_iqr < min_from_all) :
    min_iqr = min_from_all
if (max_iqr > max_from_all) :
    max_iqr = max_from_all

ti[(ti['Fare'] < min_iqr ) | (ti['Fare'] > max_iqr)] # 이상치 조회
```
